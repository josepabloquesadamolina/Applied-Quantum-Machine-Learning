# Relative error evolution

```python
iters=14

relative_errors_1= np.loadtxt('relative_errors_tot_it_14_1.csv', delimiter=',')
relative_errors_2= np.loadtxt('relative_errors_tot_it_14_2.csv', delimiter=',')
relative_errors_3= np.loadtxt('relative_errors_tot_it_14_3.csv', delimiter=',')
relative_errors_4= np.loadtxt('relative_errors_tot_it_14_4.csv', delimiter=',')
relative_errors_5= np.loadtxt('relative_errors_tot_it_14_5.csv', delimiter=',')


fig, ax = plt.subplots()
ax.plot(list(range(1,iters+1)), relative_errors_1,marker="o", linewidth=0.5, alpha=0.5, label='Simulated annealing')
ax.plot(list(range(1,iters+1)), relative_errors_2,marker="o", linewidth=0.5, alpha=0.5, label='Default annealing')
ax.plot(list(range(1,iters+1)), relative_errors_3,marker="o", linewidth=0.5,alpha=0.5, label='Reverse annealing')
ax.plot(list(range(1,iters+1)), relative_errors_4,marker="o", linewidth=0.5,alpha=0.5, label='Extended annealing')
ax.plot(list(range(1,iters+1)), relative_errors_5,marker="o", linewidth=0.5,alpha=0.5, label='Pause-Quench')
    
    
plt.legend()
ax.set(xlabel='Iterations', ylabel='|| A-BC || / || A ||', title='Relative error evolution')
ax.grid()
plt.show()
fig.savefig("Relative error evolution.png")

print("Relative error for last iteration (Simulated annealing): ",relative_errors_1[13])
print("Relative error for last iteration (Default annealing): ",relative_errors_2[13])
print("Relative error for last iteration (Reverse): ",relative_errors_3[13])
print("Relative error for last iteration (Extended annealing): ",relative_errors_4[13])
print("Relative error for last iteration (Pause-Quench): ",relative_errors_5[13])

```
